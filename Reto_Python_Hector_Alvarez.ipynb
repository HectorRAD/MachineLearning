#Data Frame 1

```python
import pandas as pd
!pip install xlsxwriter

test = pd.read_csv("/content/Test.csv")
test.head()
```

#Data Frame 2


```python
test2 = pd.read_csv("/content/Test2.csv")
test2.head()
```

#Data Frame 3


```python
train = pd.read_csv("/content/Train.csv")
train.head()
```

#Data Frame 4


```python
train2 = pd.read_csv("/content/Train2.csv")
train2.head()
```

#Data Frame Unificado

```python
tempdf = pd.concat([test, train], ignore_index=True)
tempdf2 = pd.concat([test2, train2], ignore_index=True)

columns = ["Employee_ID"]

main_df = pd.merge(tempdf, tempdf2, on="Employee_ID", how="left")

#main_df.to_csv("test3.csv")
main_df.head()
```

#Agregando 'No registrado' an entries con NaN en Sistemas operativos

```python
main_df[[" Computer_OS"," Mobile_OS"]] = main_df[[" Computer_OS"," Mobile_OS"]].fillna("No registrado")

#main_df.to_csv("test3.csv")
```

#Porcentajes

```python
#Porcentaje Computer_OS
comp_os = main_df[" Computer_OS"].value_counts(normalize=True) * 100
comp_percent = pd.DataFrame({'OS':comp_os.index, 'Percentage':comp_os.values})
comp_percent.head()

#Porcentaje Mobile_OS
mobile_os = main_df[" Mobile_OS"].value_counts(normalize=True) * 100
mobile_percent = pd.DataFrame({'OS':mobile_os.index, 'Percentage':mobile_os.values})
mobile_percent.head()
```

#Edad promedio por OS de Escritorio

```python
os_mean = main_df.groupby(' Computer_OS')['Age'].mean()
os_avg = pd.DataFrame({'OS':os_mean.index, 'Edad Promedio':os_mean.values})
os_avg.head()
```

#Tickets por OS de Escritorio

```python
tickets = main_df.groupby(' Computer_OS')[' Computer_tickets'].sum()
os_tickets = pd.DataFrame({'OS':tickets.index, 'Tickets':tickets.values})
os_tickets.head()
#print(os_tickets['OS'])
```

#Creacion de .xslx

```python
writer = pd.ExcelWriter("/content/Reto Python - Hector Alvarez.xlsx", engine="xlsxwriter")
comp_percent.to_excel(writer, sheet_name = "OS Percentage", index =False)
mobile_percent.to_excel(writer, sheet_name = "Mobile Percentage", index =False)
os_avg.to_excel(writer, sheet_name = "Edad Promedio", index =False)
os_tickets.to_excel(writer, sheet_name = "OS Tickets", index =False)
writer.close()
```
