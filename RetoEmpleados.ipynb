#Lectura de archivo


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.decomposition import PCA

EmpleadosAttrition = pd.read_csv("/content/empleadosRETO.csv")
EmpleadosAttrition.head()
```

#Eliminacion de columnas no importantes (a criterio personal)


```python
EmpleadosAttrition = EmpleadosAttrition.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], axis=1)
print(EmpleadosAttrition.columns)
```

#Creando columna YearsAtCompany

```python
EmpleadosAttrition['Year'] = EmpleadosAttrition['HiringDate'].str.split('/').str[2]
EmpleadosAttrition['YearsAtCompany'] = 2018 - EmpleadosAttrition['Year'].astype(int)

EmpleadosAttrition.head()
```

#Creando columna DistanceFromHome

```python
EmpleadosAttrition.rename(columns={'DistanceFromHome':'DistanceFromHome_km'}, inplace= True)
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.split(' ').str[0].astype(int)

EmpleadosAttrition.head()
```

#Borrando Columnas auxiliares

```python
EmpleadosAttrition = EmpleadosAttrition.drop(['Year', 'HiringDate', 'DistanceFromHome_km'], axis=1)
print(EmpleadosAttrition.columns)
```

#Sueldo promedio por departamento

```python
SueldoPromedioDeptoTemp = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean()
SueldoPromedioDepto = pd.DataFrame({'Departamento':SueldoPromedioDeptoTemp.index, 'Salario Promedio':SueldoPromedioDeptoTemp.values})
SueldoPromedioDepto.head()
```

#Escalando MonthlyIncome

```python
scaler = MinMaxScaler()

EmpleadosAttrition['MonthlyIncome'] = scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
print(EmpleadosAttrition['MonthlyIncome'])
```

##Conversion a numericas

```python
EmpleadosAttrition['BusinessTravel'] = EmpleadosAttrition['BusinessTravel'].astype('category').cat.codes
EmpleadosAttrition['Department'] = EmpleadosAttrition['Department'].astype('category').cat.codes
EmpleadosAttrition['EducationField'] = EmpleadosAttrition['EducationField'].astype('category').cat.codes
EmpleadosAttrition['Gender'] = EmpleadosAttrition['Gender'].astype('category').cat.codes
EmpleadosAttrition['JobRole'] = EmpleadosAttrition['JobRole'].astype('category').cat.codes
EmpleadosAttrition['MaritalStatus'] = EmpleadosAttrition['MaritalStatus'].astype('category').cat.codes
EmpleadosAttrition['Attrition'] = EmpleadosAttrition['Attrition'].astype('category').cat.codes
EmpleadosAttrition['OverTime'] = EmpleadosAttrition['Attrition'].astype('category').cat.codes


EmpleadosAttrition.head()
```

#Calculo de correlacion con Attrition



```python
temp = EmpleadosAttrition[['Age', 'BusinessTravel', 'Department', 'Education', 'EducationField',
       'EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel',
       'JobRole', 'JobSatisfaction', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
       'PerformanceRating', 'RelationshipSatisfaction', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsAtCompany',
       'DistanceFromHome']].corrwith(EmpleadosAttrition['Attrition'])

print(temp)
```

#Seleccionando Variables

```python
cols = []

for i, v in temp.items():
  if abs(v) >= 0.1:
    cols.append(i)
cols.append('Attrition')
print(cols)

EmpleadosAttritionFinal = pd.DataFrame()
EmpleadosAttritionFinal = EmpleadosAttrition[cols]

EmpleadosAttritionFinal.head()
```

#PCA

```python
pca = PCA()
pca.fit(EmpleadosAttritionFinal)

#print(cols)
#print(pca.explained_variance_)
#print(pca.explained_variance_ratio_)

EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
pca.fit(EmpleadosAttritionPCA)

EmpleadosAttritionFinal['C0'] = EmpleadosAttritionPCA[:,0]
EmpleadosAttritionFinal['C1'] = EmpleadosAttritionPCA[:,1]
EmpleadosAttritionFinal['C2'] = EmpleadosAttritionPCA[:,4]
EmpleadosAttritionFinal['C3'] = EmpleadosAttritionPCA[:,5]
EmpleadosAttritionFinal['C4'] = EmpleadosAttritionPCA[:,10]

EmpleadosAttritionFinal.head()
```

#Creacion de CSV

```python
EmpleadosAttritionFinal.to_csv('/content/EmpleadosAttritionFinal.csv', index=False)


#comment so it appears on the pdf
```
